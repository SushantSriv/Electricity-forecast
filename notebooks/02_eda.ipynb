```python
import pandas as pd
import matplotlib.pyplot as plt
# evt. import plotly.express as px for interaktive grafer
```

```python
# Last inn clean data
df = pd.read_csv(
    "../data/interim/clean_prices.csv",
    index_col="Unnamed: 0",
    parse_dates=["Unnamed: 0"]
)
df.index.name = "datetime"

# Sjekk form og kolonner
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())

# Plott NO1-pris
plt.figure()
plt.plot(df.index, df["NO1"])
plt.title("NO1 hourly price (NOK)")
plt.xlabel("Datetime")
plt.ylabel("Price (NOK)")
plt.tight_layout()
plt.show()
```
